```python
import csv
import pandas as pd
import numpy as np
import os
```

# Create Flatfile

```python
data = pd.read_csv('clean_data.csv', encoding = "ISO-8859-1", index_col = 0)
```

```python
countries=data.index.tolist()
values = data.values.flatten().tolist()
headers=data.columns.tolist()
clean_headers=[]
for i in range(len(headers[1:])):
    new_list = headers[i+1].split("?")
    clean_headers.append(new_list)
clean_headers[0:2]
data_indices = []
for i in range(len(countries)):
    for j in range(len(clean_headers)):
        list_1 = clean_headers[j][0:5]
        list_1.append(countries[i])
        list_1 = list_1[-1:] + list_1[:-1]
        data_indices.append(list_1)
data_indices[0:5]
for i in range(len(values)):
    data_indices[i].append(values[i])
with open("output.csv", "wb") as f:
    writer = csv.writer(f)
    writer.writerows(data_indices)
```

```python
df = pd.read_csv('output.csv', encoding = "ISO-8859-1", header=None)
df.columns =["countrs", "major_key", "key_1", "key_2", "key_3", "year", "value"]
```
